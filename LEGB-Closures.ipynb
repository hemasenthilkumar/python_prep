```python
"""
Question 1

Write a function probe() such that:

x = 10
probe()
print(x)


prints:

20


Constraints:

You are not allowed to return anything from probe

You are not allowed to reassign x outside probe
"""
x=10
def probe():
    global x
    x = 20
probe()
print(x)
```

```python
"""
Without using global, write code so that:

f = outer()
f()
f()


prints:

11
12


You may define outer however you want.
"""

def outer():
    counter = 10
    def inner():
        nonlocal counter
        counter += 1
        return counter
    return inner

f = outer()
print(f())
print(f())
```

```python
"""
Write outer() so that:

f1, f2 = outer()
f1()
f2()
f1()


prints:

1
2
3


Constraints:

f1 and f2 must share state

You may not use classes
"""

def outer():
    count = 0

    def inner1():
        nonlocal count
        count += 1
        return count
    def inner2():
        nonlocal count
        count += 1
        return count   

    return inner1, inner2


f1, f2 = outer()
print(f1())
print(f2())
print(f1())
```

```python
"""
Question 5

Implement scope_trap() such that:

x = "global"
scope_trap()


prints:

enclosing
local
global


Constraints:

Exactly three print statements

All three scopes must be involved

No global reassignment
"""

x = "global"

def scope_trap():
    y = "enclosing"
    print(y)
    def inner():
        z= 'local'
        print(z)
    inner()
    print(x)

scope_trap()
```

```python
"""
Question 3

Implement make_multipliers() such that:

funcs = make_multipliers()
print([f(10) for f in funcs])


prints:

[0, 10, 20]


Constraints:

You must use a loop

You must use a closure

Do not hardcode values
"""

def make_multipliers():
    funcs= []
    for i in range(3):
        def multiplier(x, i=i):
            return x*i
        funcs.append(multiplier)
    return funcs

funcs = make_multipliers()
print([f(10) for f in funcs])
```
